```python
import os
from lxml import etree

# Define the class names and their corresponding index
class_names = ['cats houses']
class_dict = {class_name: i for i, class_name in enumerate(class_names)}

# Define the path to the XML files and the output directory for the YOLO annotation files
xml_dir = '/content/drive/MyDrive/Labelled_images/annotations/annotations_istanbul_cat houses'
output_dir = '/content/drive/MyDrive/xmltoyolo/cats houses'

# Loop through the XML files and convert them to YOLO format
for xml_file in os.listdir(xml_dir):
    # Parse the XML file
    tree = etree.parse(os.path.join(xml_dir, xml_file))
    root = tree.getroot()
    
    # Get the image dimensions
    width = int(root.find('size/width').text)
    height = int(root.find('size/height').text)
    
    # Loop through the objects in the image and convert their coordinates to YOLO format
    objects = root.findall('object')
    yolo_lines = []
    for obj in objects:
        class_name = obj.find('name').text
        class_index = class_dict[class_name]
        xmin = float(obj.find('bndbox/xmin').text)
        xmax = float(obj.find('bndbox/xmax').text)
        ymin = float(obj.find('bndbox/ymin').text)
        ymax = float(obj.find('bndbox/ymax').text)
        x = (xmin + xmax) / 2 / width
        y = (ymin + ymax) / 2 / height
        w = (xmax - xmin) / width
        h = (ymax - ymin) / height
        yolo_lines.append(f"{class_index} {x} {y} {w} {h}")
    
    # Write the YOLO annotation file
    output_file = os.path.splitext(xml_file)[0] + '.txt'
    with open(os.path.join(output_dir, output_file), 'w') as f:
        f.write('\n'.join(yolo_lines))
```
